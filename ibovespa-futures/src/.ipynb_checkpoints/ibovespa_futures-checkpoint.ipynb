# Relatório sobre contratos futuros do ibovespa

Neste relatório será apresentado insights sobre um dataset de contratos futuros do ibovespa capturado no período
entre 19/07/2019 a 14/08/2019.

O dataset possui registros categorizados por data, hora, valor de abertura, valor de fechamento, valor máximo e
valor mínimo.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("winq19.csv")
df
```

### Valor de abertura x fechamento do índice a cada dia

```python
plt.figure(1, figsize=(20,10))
fig, ax = plt.subplots()
plt.ylabel('Valor do índice')
plt.xlabel('Dia de captura do índice')

serie1 = df.sort_values(by=['date', 'time'])\
.groupby('date')\
.head(1)\
.plot(x='date', y='open', ax=ax, color='blue', grid=True)

serie2 = df.sort_values(by=['date', 'time'])\
.groupby('date')\
.tail(1)\
.plot(x='date', y='close', ax=ax, color='red', grid=True)

serie1.legend(loc=1)
serie2.legend(loc=2)

plt.show()
```

### Valor máximo x mínimo do índice a cada dia

```python
plt.figure(1, figsize=(20,10))
fig, ax = plt.subplots()
plt.ylabel('Valor do índice')
plt.xlabel('Dia de captura do índice')

serie1 = df.loc[df.groupby('date').low.idxmin()]\
.plot(x='date', y='low', ax=ax, color='blue', grid=True)

serie2 = df.loc[df.groupby('date').high.idxmax()]\
.plot(x='date', y='high', ax=ax, color='red', grid=True)

serie1.legend(loc=1)
serie2.legend(loc=2)

plt.show()
```
